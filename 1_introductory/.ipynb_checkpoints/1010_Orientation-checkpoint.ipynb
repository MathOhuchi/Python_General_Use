# Ⅰ．社会人のためのPython活用術・入門


## 1. 社会人のためのPython

### 社会人のためのPythonとは
昨今，人工知能や機械学習の素晴らしい成果がニュースにならない日が無いくらいブームになっています．そのプログラムの多くはPythonで書かれているため，自ずとPythonへの注目度も上がり，Pythonへの学習熱も盛んになっています．多くの人は，Pythonは人工知能などの高度なプログラムを構築するためのプログラム言語だと思っています．
しかし，普通の社会人でもPythonが使えて一般業務へ活用することが大変有効であることが認知されていません．Pythonは自分とは関係のないものと思っているかもしれませんが，実はそんなに難しくない使い方で一般業務が劇的に変わります．そのきっかけとなった技術がJupyter notebookです．これはPythonの開発環境の一つでウェブブラウザーの上で使用する協力なツールです．Jupyter notebookを使うことによって，オフィス業務の中でも非効率で個人スキルへの依存度の高い非定形業務をドキュメント化しながら高品質な仕事に変貌させることができます．

- Pythonの簡単な使い方はすぐマスター
- Jupyter notebookで業務改善
- 個人のデータ処理分析能力が着実に成長

かつてプログラム言語Perlの熟練した使い手が，数日かかるデータ処理を数時間で終わたというような伝説がありました．プログラム言語にはそのような可能性が秘められています．かつてPerlプログラマーが書くプログラムは一般人には解読不能でした．今でもその流れはハッカー達に受け継がれています．しかし，PythonとJupyter notebookの出現によって業務ハッキングが平凡で標準的な技術になりました．

毎週のように同じようなデータをExcelで処理している人は沢山います．毎回，同じような処理なのですが，ほんの少しづつ変化しているので自動化できない上，他人に任せられない状況になっています．そのような仕事はいずれ人工知能が肩代わりしてくれるでしょうが，それまでは人間が対応しなければなりません．できれば，そのような仕事にかかる時間を短縮したいものです．

### 前提スキル
前提となるスキルのハードルは高くありません．現代社会はIT化社会なので，一般企業で働く人の多くはパソコンを使っていることでしょう．パソコンを使っていない人にとっては不要な事かもしてません．きっと，もっと素晴らしいことに時間を割いているのでしょう．残念ながら日々パソコンに向かって仕事をしている人は，前提スキルに達していることでしょう．

- パソコンの操作ができる
- ウェブブラウザーが使える
- プログラムというものを大まかに知っている

さらに次のことができれは，間違いなくPythonとJupyter notebookの世界に飛び込むべきです．
- Excelなどの表計算ソフトや文書作成ソフトなどを使っている
- Visual Basiicなどでプログラムを作ったことがある
