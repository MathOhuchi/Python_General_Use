## 5. マークダウンの使い方
Jupyter notebookの入力可能な部分をセルと言います．セルにはCode, Markdown, Raw NBConvertの3種類ありますが，デフォルトはCodeセルになっています．
セルのタイプは上部メニューによりMarkdownに変更できます．Markdownセルは，文章や数式をマークダウンの要領で記載する部分です．マークダウンに記載された内容は，プログラムの実行には一切関与しません．したがって高機能のコメントとして位置づけられます．

![Markdownセルへの変更](./images/notebook14.jpg)


## notebook内容の構成
一つのnotebookは，拡張子ipynbを持つファイルになっています．Jupyter notebookの目的はPythonのプログラムを作成して実行することですが，そのプログラムを使う仕事を包括的に文書化することがJupyter notebookを利用する最大のメリットになります．私たちは，このメリットを最大限に活かして効率的に業務を遂行する作法を獲得します．Jupyter notebookのCodeセルにプログラムを記載して実行したならば，その前後にMarkdownセルを作り，Codeセルの内容説明と結果についての評価を記載しましょう．
#### セルの構成
一つのnotebookは複数のCodeセルに分割されるので，それらの前後にMarkdownセルを挿入して以下の構造を繰り返すように構成する．
- これから実施するCodeセルの内容の説明
- プログラムを記述したCodeセル
- 実行結果についての考察

あなたは既にここまでの学習においてMarkdownセルの恩恵を享受しています．Codeセルの前後の文章は全てMarkdownセルによって記述されています．
その記法さえ知っていれば，簡単にMarkdownを記述できるようになります．

## よく使うMarkdown
Markdownセルの中では，主に次の記述で構成されています．

- 通常の文
- 見出し
- 引用文
- 表
- 数式TeX
- Pythonプログラム例
- 箇条書き
- htmlタグの埋め込み

通常の文については，何も特別なことはなく，Markdownセルに文字テキストを記入していくだけです．
ただし，Markdownには特殊な役割をする文字があります．例えば「\*Itaric\*」は*Itaric*と表示され，「\*\*太文字\*\*」は**太文字**と表示されます．
単純にアスタリスク「\*」を表示させられないので，その場合はアスタリスクの前にバックスラッシュを置いて「\\\*」と記載します．
バックスラッシュについても同様で，単純にバックスラッシュを表示するには「\\\\」と2個並べます．バックスラッシュを付けた表記法をエスケープシーケンスと言います．
*****

### 見出し
見出しを記述するには，先頭に「#」を付けます．セルを実行することによって見出しに変換されます．
正に，ここで記載した「見出し」は次のように書かれています．
```
### 見出し
```
文の先頭に#を記載する見出しの書き方は，#の個数を増やしていくと段々と小見出しになっていきます．以下に，#の個数を一つずつ増やした例を表示します．
```
# 見出し１（#1個）
## 見出し２（#2個）
### 見出し３（#3個）
#### 見出し４（#4個）
##### 見出し５（#5個）
```
この記述が変換されると下記の表示になります．
# 見出し１（#1個）
## 見出し２（#2個）
### 見出し３（#3個）
#### 見出し４（#4個）
##### 見出し５（#5個）

運用としては分かりやすい文章を作るために，適度に見出しを挿入しましょう．
*****

### 引用文
文章の中で別の文章を引用する場合，その文が本文とは別のものであることを示すために文の左側に縦棒を付けて明示します．

> この文は引用文です．

この文は文の左に「>」を付けて記載しています．
```
> この文は引用文です．
```

このように記載して実行ボタンをクリックすると引用文の表現に変換されます．
*****

### 表形式
文章の中に表を埋め込むと全体の説明力が向上します．
次に表形式のレイアウトについて確認しましょう．

| 二項演算          | Python記法 |
| :---              | :---:      |
| 加法 $a+b$        | a + b      |
| 減法 $a-b$        | a - b      |
| 乗法 $a \times b$ | a * b      |
| 除法 $a \div b$   | a / b      |

この表は，マークダウンに次にように記述されており，実行ボタンをクリックすることによって表に変換されます．

```
| 二項演算          | Python記法 |
| :---              | :---:      |
| 加法 $a+b$        | a + b      |
| 減法 $a-b$        | a - b      |
| 乗法 $a \times b$ | a * b      |
| 除法 $a \div b$   | a / b      |
```

この表の記述の2行目は表中の文字列を整列条件を明示しています．
「 :---: 」は中央揃え，「:--- 」は左寄せ，「 ---: 」は右寄せとなります．
この表中にはTeX の記述もあります．
*****

### TeX数式
MarkdownセルにはTeXのマークアップを記載することができます．TeXのマークアップを「\$」で囲めば行内での数式に変換されます．独立した行としてのTeXは「\$\$」で囲んで表現します．
```
$(2 + 3)\times(4 - 7)\div2$
```
この行は，TeXの記載方法によって記述されています．実行ボタンをクリックすると，これが変換されて，

$(2 + 3)\times(4 - 7)\div2$

となります．TeXでは掛け算記号を「\\times」で記載し，割り算記号を「\\dev」と記載します．TeXの記載方法については，TeXの専門書を参照してください．
*****

### Pythonプログラム例
PythonコードをCodeセルへの記載と全く同じように見せる場合は，バックコート「\`」を3個並べた行で前後を挟みます．さらに前の3連バックコートの後に「Python」と記載することにより，Codeセルの表記と同じ彩色となります．

```Python
(2+3)*(4-7)/2
```

というように，Jupyter notebookで見るとPythonの記法として意味のある文字に色が付きます．（ただし，htmlに搬出してブラウザーで見た場合は色付けが無くなります．）
Markdownに記載する原文は次のようになっています．
```
□□□Python
(2+3)*(4-7)/2
□□□
```

ここで四角形□はバックコート「\`」で置き換えてください
*****

### 箇条書きリスト
箇条書きは文章の中で頻繁に使用されます．箇条書きには番号付きと番号無しがあります．
箇条書きの例を見てみましょう．

1. 形の区分
    - 円形
    - 四角形
    - 三角形
- 色の区分
    - 赤色
    - 青色
    - 緑色
- 大きさの区分
    1. 小さい
    - 中くらい
    - 大きい

このように，箇条書きは入れ子にすることができ，番号付きと番号無しの箇条書きを使うことができます．
この例の箇条書きについて，実際の書き方を確認しましょう．
```
1. 形の区分
    - 円形
    - 四角形
    - 三角形
- 色の区分
    - 赤色
    - 青色
    - 緑色
- 大きさの区分
    1. 小さい
    - 中くらい
    - 大きい
```
このようにとてもシンプルな記述法になっています．
*****

### htmlタグの埋め込み
Markdownは簡単に整形された文章を書くための道具ですが，あまり凝ったことはできません。
しかしMarkdownにはhtmlのタグを埋め込むことができるので，細かな表現をhtmlでかばーすることが可能になっています．

一つの例として，文中の<font color=red>文字の色</font>を赤に変えてみます．

この文中で赤いフォントになってる部分は「``<font color=red>文字の色</font>``」と記載されています．
